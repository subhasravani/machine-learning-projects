```python
titanic.drop('Cabin',axis=1,inplace=True)
titanic
```

```python
titanic.dropna(inplace=True)
titanic.isnull().sum()
```

```python
sex=pd.get_dummies(titanic['Sex'])
sex.head()
```

```python
embark=pd.get_dummies(titanic['Embarked'])
embark.head()
```

```python
pcl=pd.get_dummies(titanic['Pclass'],drop_first=True)
pcl.head()
```

```python
titanic = pd.concat([titanic,sex,embark,pcl],axis=1)
titanic.head()
```

```python
titanic.drop(['Name','Pclass','Embarked','Sex','Ticket'],axis=1,inplace=True)
```

```python
titanic
```

```python
x=titanic.drop('Survived',axis=1)
y=titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
accuracy=confusion_matrix(y_test,y_pred)
accuracy
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
titanic=pd.DataFrame({"Actual":y_test,'predicted':y_pred})
sns.regplot(x='Actual',y='predicted',data=titanic,color='green')
plt.figure(figsize=(20,8))
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
